```python
import pandas as pd
import numpy as np
import os, sys
from IPython.display import display
from IPython.display import Image as _Imgdis
from numpy import asarray
from matplotlib import pyplot

df1 = pd.read_csv("Bittrex_BTCUSD_1h.csv", header=1)
df2 = pd.read_csv("Bittrex_ETHUSD_1h.csv",header=1)
df3 = pd.read_csv("Bittrex_LTCUSD_1h.csv",header=1)


df1["open_change"] = (df1["Open"] / df1["Open"].shift(-1)) -1
df1["high_change"] = (df1["High"] / df1["High"].shift(-1)) -1
df1["low_change"] = (df1["Low"] / df1["Low"].shift(-1)) -1
df1["close_change"] = (df1["Close"] / df1["Close"].shift(-1)) -1

df2["open_change"] = (df2["Open"] / df2["Open"].shift(-1)) -1
df2["high_change"] = (df2["High"] / df2["High"].shift(-1)) -1
df2["low_change"] = (df2["Low"] / df2["Low"].shift(-1)) -1
df2["close_change"] = (df2["Close"] / df2["Close"].shift(-1)) -1

df3["open_change"] = (df3["Open"] / df3["Open"].shift(-1)) -1
df3["high_change"] = (df3["High"] / df3["High"].shift(-1)) -1
df3["low_change"] = (df3["Low"] / df3["Low"].shift(-1)) -1
df3["close_change"] = (df3["Close"] / df3["Close"].shift(-1)) -1

df1 = df1.drop("Unix Timestamp", axis=1)
df1 = df1.drop("Volume BTC", axis=1)
df1 = df1.drop("Volume USD", axis=1)
df1.drop(df1.tail(3).index,inplace=True)

df2 = df2.drop("Unix Timestamp", axis=1)
df2 = df2.drop("Volume ETH", axis=1)
df2 = df2.drop("Volume USD", axis=1)
df2.drop(df2.tail(3).index,inplace=True)

df3 = df3.drop("Unix Timestamp", axis=1)
df3 = df3.drop("Volume LTC", axis=1)
df3 = df3.drop("Volume USD", axis=1)
df3.drop(df3.tail(3).index,inplace=True)




df1 = df1.dropna()
df2 = df2.dropna()
df3 = df3.dropna()

df = pd.concat([df1, df2, df3])
df
```

```python

li = [len(df3.index), len(df2.index), len(df3.index), len(df1.index) ]
curr_min = 999999
for i in range(0, len(li)):
    if li[i] < curr_min:
        curr_min = li[i]
X = []
Y = []

changes = df[["open_change", "high_change", "low_change", "close_change"]].to_numpy()


for i in range(0,  curr_min): 
    Y.append(changes[i])  
    X.append(np.array(changes[i+1:i+21][::-1]))


X = np.array(X).reshape(-1, 20, 4)
Y = np.array(Y)
```

```python
from keras.models import Sequential 
from keras.layers import LSTM

model = Sequential()
model.add(LSTM(1, input_shape=(20, 4)))

model.compile(optimizer="rmsprop", loss="mse")
model.fit(X, Y, batch_size=32, epochs=500)
```
